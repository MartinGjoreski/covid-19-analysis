```python
api_key1 = "368776ee77a2a95e025f72b962067a7f" #martin.gjoreski@gmail.com
api_key2 = "b756b1871d9a7d030fdd7776791ef9f5" #martin.gjoreski@gmail.com
api_key3 = "9389332caab7b9ee6faadfcd023e386a" #martin.gjoreski@ijs.si
api_key4 = "8a44d2add848ce91b0517c6f8c8120c4" #martin_gjoreski@hotmail.com
api_key5 = "c97730784a22fbbe6912c4cd8f1faee1" #martin_gjoreski@yahoo.com
api_key6 = "41a8bd0dfa8023c479a548d8af4aeb9c" #martin.gjoreski@fri.uni-lj.si


#for each country and each date ger weather data
start_row = 34
api_key=api_key6
for row_idx in range(len(df)):
    if row_idx<start_row:
        continue
    row_data = df.iloc[row_idx]
    df_country_arr = []

    latitude = df.Lat.iloc[row_idx]
    longitude = df.Long.iloc[row_idx]
    
    for date_idx in range(4,len(df.columns[4:])):
        date = df.columns[date_idx]
        date = date+'20'
        timestamp = time.mktime(datetime.datetime.strptime(date, "%m/%d/%Y").timetuple())
        timestamp  = timestamp+12*60*60
        timestamp
        # Create URL template
        url_template = "https://api.darksky.net/forecast/{}/{},{},{}"

        # Create request URL
        request_url = url_template.format(
            api_key,
            latitude,
            longitude,int(timestamp))

        print(request_url)
        #Get historical weather data as JSON 
        response = requests.get(request_url)

        # Handle unsuccessful response code
        if not response.ok:
            print(response.json())
            break
            continue

        # Get JSON response
        json_data = response.json()

        # Get daily data from JSON data
        daily_data = json_data["daily"]["data"][0]

        df_tmp = pd.DataFrame([list(daily_data.values())],columns = list(daily_data.keys()))
        df_tmp['Province/State'] = row_data['Province/State']
        df_tmp['Country/Region'] = row_data['Country/Region']
        df_tmp['Lat'] = row_data['Lat']
        df_tmp['Long'] = row_data['Long']
        columns = df_tmp.columns.values
        columns = np.concatenate([columns[-4:],columns[:-4],])
        df_tmp = df_tmp[columns]
        df_tmp = df_tmp.T
        df_tmp.columns = [date]
        df_country_arr.append(df_tmp) 
        
    df_tmp  = pd.concat(df_country_arr,axis=1)
    country = df_tmp.loc['Country/Region'][0]
    last_date = df_tmp.columns[-1].replace('/','-')
    Province_State = df_tmp.loc['Province/State'][0]
    if "nan"  in str(Province_State):
        Province_State = country
    name = country+"-"+Province_State+"-"+last_date+".csv"
    df_tmp.to_csv('weather/'+name)
    
    print('Done with',name,row_idx)
```
